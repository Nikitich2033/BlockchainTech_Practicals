Like the transaction ID, the block ID is the hex representation of the hash256 of the header The block ID is interesting: this ID is what gets put into prev_block for a block building on top of this one. 

````python
from helper import hash256
block_hash = hash256(bytes.fromhex('020000208ec39428b17323fa0ddec8e887b4a7c53b8c0a0a220cfd0000000000000000005b0750fce0a889502d40508d39576821155e9c9e3f5c3157f961db38fd8b25be1e77a759e93c0118a4ffd71d'))[::-1]
block_id = block_hash.hex()
print(block_id) 
````

Proof-of-work is the requirement that the hash of every block header in Bitcoin must be below a certain target. The target is a 256-bit number that is computed directly from the bits field (in our example, e93c0118). 

````python
 from helper import little_endian_to_int
 bits = bytes.fromhex('e93c0118')
 exponent = bits[-1]
 coefficient = little_endian_to_int(bits[:-1])
 target = coefficient * 256**(exponent - 3)
 print('{:x}'.format(target).zfill(64))
````

We can check that this block header’s hash satisfies the proof-of-work as follows:

````python
from helper import little_endian_to_int
proof = little_endian_to_int(hash256(bytes.fromhex('020000208ec39428b17323fa0ddec8e887b4a7c53b8c0a0a220cfd0000000000000000005b0750fce0a889502d40508d39576821155e9c9e3f5c3157f961db38fd8b25be1e77a759e93c0118a4ffd71d')))
print(proof < target)
True
````

Targets are hard for human beings to comprehend. The target is the number that the hash must be below, but as humans, it’s not easy to see the difference between a 180- bit number and a 190-bit number. The first is a thousand times smaller, but from looking at targets, such large numbers are not easy to contextualize.
To make different targets easier to compare, the concept of difficulty was born. The trick is that difficulty is inversely proportional to the target, to make comparisons easier.

````python
from helper import little_endian_to_int
bits = bytes.fromhex('e93c0118')
exponent = bits[-1]
coefficient = little_endian_to_int(bits[:-1]) 
target = coefficient*256**(exponent-3)
difficulty = 0xffff * 256**(0x1d-3) / target 
print(difficulty)

````

Send BTC via Testnet

The construction of a transaction requires answering some basic questions:
1. Where do we want the bitcoins to go?
2. What UTXOs can we spend?
3. How quickly do we want this transaction to get into the blockchain?

1- Create (a couple) of addresses Testnet-compatible
2- Ask for some testnet BTC from faucets
3- Compile the transaction
4-Sign it
5- Send it

````python
from ecc import PrivateKey
from helper import hash256, little_endian_to_int
secret = little_endian_to_int(hash256(b'Amy02'))
private_key = PrivateKey(secret)
print(private_key.point.address(testnet=True))
````

````python
from helper import decode_base58, SIGHASH_ALL
from script import p2pkh_script, Script
from tx import TxIn, TxOut, Tx
prev_tx = bytes.fromhex('3e003760c2a5df3827471a28d17dd8d3523cb3b597617b7147147b6d8051f85b')
prev_index = 1
tx_in = TxIn(prev_tx, prev_index)
tx_outs = []
change_amount = int(10)
change_h160 = decode_base58('mnrVtF8DWjMu839VW3rBfgYaAfKk8983Xf')
change_script = p2pkh_script(change_h160)
change_output = TxOut(amount=change_amount, script_pubkey=change_script)
target_amount = int(10000)
target_h160 = decode_base58('mjthyg9bidchj8sEMZxQxTXTpKgcrZhnw7')
target_script = p2pkh_script(target_h160)
target_output = TxOut(amount=target_amount, script_pubkey=target_script)
tx_obj = Tx(1, [tx_in], [change_output, target_output], 0, True)
print(tx_obj)
````

````python
from ecc import PrivateKey
from helper import SIGHASH_ALL
z = tx_obj.sig_hash(0)
private_key = PrivateKey(secret=30289000167393822809387466441492771014912117484804967039633807558721972070656)
der = private_key.sign(z).der()
sig = der + SIGHASH_ALL.to_bytes(1, 'big')
sec = private_key.point.sec()
script_sig = Script([sig, sec])
tx_obj.tx_ins[0].script_sig = script_sig 
print(tx_obj.serialize().hex())
````

Merkle tree structure

````python
import math
total = 27
max_depth = math.ceil(math.log(total, 2))
merkle_tree = []
for depth in range(max_depth + 1):
        num_items = math.ceil(total / 2**(max_depth - depth))
        level_hashes = [None] * num_items 
        merkle_tree.append(level_hashes)
for level in merkle_tree:
        print(level)
````
